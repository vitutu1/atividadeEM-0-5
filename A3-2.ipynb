```python
#1 soma

def SomaPoli(poly1, poly2, index=0):
    if index >= len(poly1) and index >= len(poly2):
        return []
    
    coef1 = poly1[index] if index < len(poly1) else 0
    coef2 = poly2[index] if index < len(poly2) else 0
    
    return [coef1 + coef2] + SomaPoli(poly1, poly2, index + 1)

def GravarPoli():
    grau = int(input("Informe o grau do polinômio: "))
    polinomio = []
    for i in range(grau + 1):
        coef = int(input(f"Informe o coeficiente para x^{i}: "))
        polinomio.append(coef)
    return polinomio

def main():
    NumeroPoli = int(input("Quantos polinomios deseja somar? "))
    polinomios = []

    for i in range(NumeroPoli):
        print(f"\nDigite os numeros para o polinomio {i + 1}:")
        polinomios.append(GravarPoli())

    resultado = polinomios[0]
    for i in range(1, NumeroPoli):
        resultado = SomaPoli(resultado, polinomios[i])

    print("\nResultado da soma dos polinomios:")
    for i in range(len(resultado)):
        print(f"{resultado[i]}x^{i}", end=" ")
        if i < len(resultado) - 1:
            print("+", end=" ")
    print()

main()
```

```python
#2

def SubPoli(poly1, poly2, index=0):
    if index >= len(poly1) and index >= len(poly2):
        return []
    
    coef1 = poly1[index] if index < len(poly1) else 0
    coef2 = poly2[index] if index < len(poly2) else 0
    
    return [coef1 - coef2] + SubPoli(poly1, poly2, index + 1)

def GravarPoli():
    grau = int(input("Informe o grau do polinomio: "))
    polinomio = []
    for i in range(grau + 1):
        coef = int(input(f"Informe o numero para x^{i}: "))
        polinomio.append(coef)
    return polinomio

def main():
    NumeroPoli = int(input("Quantos polinomios deseja subtrair? "))
    polinomios = []

    for i in range(NumeroPoli):
        print(f"\nDigite os coeficientes para o polinomio {i + 1}:")
        polinomios.append(GravarPoli())

    resultado = polinomios[0]
    for i in range(1, NumeroPoli):
        resultado = SubPoli(resultado, polinomios[i])

    print("\nResultado da subtracao dos polinomios:")
    for i in range(len(resultado)):
        print(f"{resultado[i]}x^{i}", end=" ")
        if i < len(resultado) - 1:
            print("-" if resultado[i+1] >= 0 else "+", end=" ")
    print()

main()
```

```python
#3 multi
def criar_polinomio():
    coef2 = int(input("Coeficiente de x² (ou 0 se não houver termo): "))
    coef1 = int(input("Coeficiente de x: "))
    const = int(input("Constante: "))
    return [(coef2, 2), (coef1, 1), (const, 0)]

def multiplicar_termos(termo1, termo2):
    coef1, exp1 = termo1
    coef2, exp2 = termo2
    novo_coef = coef1 * coef2
    novo_exp = exp1 + exp2
    return novo_coef, novo_exp

def multiplicar_polinomios(p1, p2):
    if not p1:
        return []
    termo1, *resto_p1 = p1
    resultado_parcial = [multiplicar_termos(termo1, termo2) for termo2 in p2]
    return resultado_parcial + multiplicar_polinomios(resto_p1, p2)

def simplificar_polinomio(polinomio):
    resultado = {}
    for coef, exp in polinomio:
        if exp in resultado:
            resultado[exp] += coef
        else:
            resultado[exp] = coef
    return [(coef, exp) for exp, coef in sorted(resultado.items(), reverse=True) if coef != 0]

# Exemplo de uso
print(f"Polinômio 1: {p1}")
p1 = criar_polinomio()
print(f"Polinômio 2: {p2}")
p2 = criar_polinomio()
resultado = multiplicar_polinomios(p1, p2)
resultado_simplificado = simplificar_polinomio(resultado)
print("Resultado simplificado:", resultado_simplificado)

```

```python
#4 dividir

def criar_polinomio():
    coef2 = int(input("Coeficiente de x² (ou 0 se não houver termo): "))
    coef1 = int(input("Coeficiente de x: "))
    const = int(input("Constante: "))
    return [(coef2, 2), (coef1, 1), (const, 0)]

def dividir_polinomios(p1, p2):
    quociente = []
    resto = p1
    
    while resto and (resto[0][1] >= p2[0][1]):
       
        coef_quociente = resto[0][0] // p2[0][0]
        exp_quociente = resto[0][1] - p2[0][1]
        quociente.append((coef_quociente, exp_quociente))
        
        
        polinomio_subtrair = [(coef_quociente * coef, exp_quociente + exp) for coef, exp in p2]
        
        
        novo_resto = []
        i, j = 0, 0
        
        while i < len(resto) or j < len(polinomio_subtrair):
            if i < len(resto) and (j >= len(polinomio_subtrair) or resto[i][1] > polinomio_subtrair[j][1]):
                novo_resto.append(resto[i])
                i += 1
            elif j < len(polinomio_subtrair) and (i >= len(resto) or resto[i][1] < polinomio_subtrair[j][1]):
                novo_resto.append((-polinomio_subtrair[j][0], polinomio_subtrair[j][1]))
                j += 1
            else:  # caso de termos com o mesmo grau
                coef_novo = resto[i][0] - polinomio_subtrair[j][0]
                if coef_novo != 0:
                    novo_resto.append((coef_novo, resto[i][1]))
                i += 1
                j += 1
        
        resto = novo_resto
        
    return quociente, resto


print(f"Polinômio 1: {p1}")
p1 = criar_polinomio()
print(f"Polinômio 2: {p2}")
p2 = criar_polinomio()
quociente, resto = dividir_polinomios(p1, p2)

print("Quociente:", quociente)
print("Resto:", resto)


```

```python
#5
def fator_comum_em_evidencia(polinomio):
    fator_comum = min([coef for coef, _ in polinomio if coef != 0])
    fatorado = [(coef // fator_comum, exp) for coef, exp in polinomio]
    return fatorado, fator_comum


polinomio = criar_polinomio()
fatorado, fator_comum = fator_comum_em_evidencia(polinomio)
print("Polinômio fatorado:", fatorado)
print("Fator comum:", fator_comum)
```

```python
def trinômio_quadrado_perfeito(polinomio):
    coef2, coef1, const = polinomio[0][0], polinomio[1][0], polinomio[2][0]
    if coef1 ** 2 == 4 * coef2 * const:
        raiz = int(coef1 / (2 * coef2))
        return f"(x + {raiz})^2"
    else:
        return "Não é um trinômio quadrado perfeito."


polinomio = criar_polinomio()
fatorado = trinômio_quadrado_perfeito(polinomio)
print("Fatoração como trinômio quadrado perfeito:", fatorado)
```

```python
def diferenca_dois_quadrados(polinomio):
    if len(polinomio) != 2:
        return "Não é uma diferença de dois quadrados."
    coef1, exp1 = polinomio[0]
    coef2, exp2 = polinomio[1]
    if exp1 == 2 and exp2 == 0 and coef1 > 0 and coef2 < 0:
        raiz1 = int(coef1 ** 0.5)
        raiz2 = int((-coef2) ** 0.5)
        return f"(x + {raiz2})(x - {raiz2})"
    else:
        return "Não é uma diferença de dois quadrados."


polinomio = criar_polinomio()
fatorado = diferenca_dois_quadrados(polinomio)
print("Fatoração como diferença de dois quadrados:", fatorado)
```

```python
def soma_dois_cubos(polinomio):
    if len(polinomio) == 2 and polinomio[0][1] == 3 and polinomio[1][1] == 0:
        a = int(polinomio[0][0] ** (1/3))
        b = int(polinomio[1][0] ** (1/3))
        return f"({a} + {b})(x^2 - {a}x + {b})"
    return "Não é uma soma de dois cubos."

def diferenca_dois_cubos(polinomio):
    if len(polinomio) == 2 and polinomio[0][1] == 3 and polinomio[1][1] == 0:
        a = int(polinomio[0][0] ** (1/3))
        b = int(-polinomio[1][0] ** (1/3))
        return f"({a} - {b})(x^2 + {a}x + {b})"
    return "Não é uma diferença de dois cubos."


polinomio = criar_polinomio()
fatorado_soma = soma_dois_cubos(polinomio)
print("Fatoração como soma de dois cubos:", fatorado_soma)
fatorado_diferenca = diferenca_dois_cubos(polinomio)
print("Fatoração como diferença de dois cubos:", fatorado_diferenca)
```
